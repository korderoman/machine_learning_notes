```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Par치metros
vocab_size = 10        # Simulamos un vocabulario peque침o con 10 palabras
embedding_dim = 4      # Cada palabra se representar치 con un vector de tama침o 4

# Crear capa de embedding
embedding = nn.Embedding(vocab_size, embedding_dim)

# Simular una secuencia de 5 palabras (por 칤ndice)
example_sequence = torch.tensor([2, 5, 0, 3, 8])  # 칤ndices en el vocabulario

# Obtener vectores de embedding
embedded = embedding(example_sequence)  # Shape: [5, 4]

# Mostrar valores de embedding
print("游댝 Secuencia de 칤ndices:")
print(example_sequence)
print("\n游닄 Vectores embebidos (una fila por palabra):")
print(embedded)

# Visualizar cada dimensi칩n del embedding
plt.figure(figsize=(8, 4))
for i in range(embedding_dim):
    plt.plot(embedded[:, i].detach().numpy(), label=f'Dim {i}')
plt.title("Representaci칩n embebida de la secuencia")
plt.xlabel("Posici칩n en la secuencia")
plt.ylabel("Valor del embedding")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
