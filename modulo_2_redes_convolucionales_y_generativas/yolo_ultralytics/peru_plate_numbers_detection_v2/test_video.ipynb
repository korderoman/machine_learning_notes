# Importamos el modelo guardado

```python
from ultralytics import YOLO
model = YOLO("runs/detect//yolo_car_detector2/weights/best.pt")
```

```python
path_video=r"D:\programacion\proyectos\utec\IA\modulo_2_redes_convolucionales_y_generativas\yolo_ultralytics\peru_plate_numbers_detection_v2\assets\demo2.mp4"
```

```python
results = model(path_video, stream=True, save=True)
```

```python
#ver el video
import cv2

# Abrir video
cap = cv2.VideoCapture(path_video)
# === 3. Obtener tamaño del frame y FPS ===
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# === 4. Crear el objeto para guardar video de salida ===
out = cv2.VideoWriter(
    "salida.mp4",                               # nombre de salida
    cv2.VideoWriter_fourcc(*'mp4v'),            # códec
    fps,                                        # frames por segundo
    (width, height)                             # tamaño del video
)

# === 5. Procesar cada frame ===
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # === 6. Detección con YOLO ===
    results = model(frame)

    # === 7. Dibujar las detecciones ===
    annotated_frame = results[0].plot()

    # === 8. Mostrar en ventana (opcional) ===
    cv2.imshow("Detección YOLO", annotated_frame)

    # === 9. Guardar el frame al nuevo video ===
    out.write(annotated_frame)

    # === 10. Salir con 'q' ===
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# === 11. Liberar todo ===
cap.release()
out.release()
cv2.destroyAllWindows()
```
